```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel('Thw Kiel comparison.xlsx')
```

```python
data
```

```python
luk_nil = data[(data['name'] == 'Lucas Nilsson') & (data['ligue'] == 'sum')]
luk_nil
```

```python
dom_duv = data[(data['name'] == 'Domagoj Duvnjak') & (data['ligue'] == 'sum')]
dom_duv
```

```python
nik_bil = data[(data['name'] == 'Nikola Bilyk') & (data['ligue'] == 'sum')]
nik_bil
```

```python
plt.style.available
plt.style.use('ggplot')
```

```python
y1 = list(luk_nil['goals'])
y2 = list(dom_duv['goals'])
x = luk_nil['season']
y1
y2
```

```python
y1
```

```python
import numpy as np
```

```python
plt.plot(x, y1, linewidth=3, color='blue', label='Lucas Nilsson')
plt.plot(x, y2, linewidth=3, color='grey', label='Domagoj Duvnjak', linestyle='--')
z1 = np.array(y1)
z2 = np.array(y2)
plt.fill_between(x, y1, y2, where=(z1 >= z2), interpolate=True, color='green')
plt.fill_between(x, y1, y2, where=(z1 < z2), interpolate=True, color='red')
plt.legend(loc='lower right')
plt.xlabel('Season')
plt.ylabel('Goals')
```
