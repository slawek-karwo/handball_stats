```python
import pandas as pd
```

```python
data = pd.read_excel('dane_superliga.xlsx')
```

```python
data.head(8)
```

```python
data[(data['home'] == 'Azoty Tarnów') | (data['away'] == 'Azoty Tarnów')]
```

```python
home = data[data['home'] == 'Azoty Tarnów']
```

```python
home.head(5)
```

```python
away = data[data['away'] == 'Azoty Tarnów']
```

```python
away.head(5)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
away
```

```python
home
```

```python
joined_df = pd.merge(home, away, how='outer', left_on='away', right_on='home')
```

```python
joined_df
```

```python
joined_df.home_y.fillna(joined_df.away_x, inplace=True)
joined_df.fillna(0, inplace=True)
```

```python
joined_df
```

```python
joined_df['home_res'] = joined_df['home_score_x'] - joined_df['away_score_x']
joined_df['away_res'] = joined_df['away_score_y'] - joined_df['home_score_y']
joined_df['bilans'] = joined_df['home_res'] + joined_df['away_res']
```

```python
joined_df
```

```python
import numpy as np
```

```python
x_ind = np.arange(len(joined_df['home_y']))
```

```python
x_ind
```

```python
width = 0.4
plt.bar(x_ind - width/2, joined_df['home_res'], width=width, label='Home')
plt.bar(x_ind + width/2, joined_df['away_res'], width=width, label='Away')
plt.plot(x_ind, joined_df['bilans'], marker='o', linestyle='None', color='black', label='Sum of goals difference')
plt.tick_params(axis = 'x', which = 'major', labelsize = 8, rotation = 80)
plt.tick_params(axis = 'y', which = 'major', labelsize = 10)
plt.xticks(ticks=x_ind, labels=joined_df['home_y'])
plt.xlabel('Opponent', size='small')
plt.ylabel('Goals Difference', size='small')
plt.legend(prop={'size': 8})
plt.title('Grupa Azoty Tarnów - season 2019/2020', size='medium')
```
